#### What percentage of all licensed sets ever released were star wars themed?

```python
merged = df.merge(theme, left_on = 'parent_theme', right_on = 'name')
merged.drop(columns = 'name_y', inplace = True)
merged
```

```python
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset = ['set_num'])
licensed

star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
star_wars.head()

#tipo = licensed['parent_theme'].str.startswith('Star Wars').sum()
#tipa = len(licensed)

the_force = int((star_wars.shape[0]/licensed.shape[0])*100)
print(the_force)
```

#### How many unique sets were released each year (1955 - 2017)?

```python
# breaking down the number of sets released each year

merged.sort_values('year', ascending = True)
```

```python
# understanding the unique type sets returing all the unique ones with counting
clean_df = merged[~merged['set_num'].isnull()]
clean_df

clean_df['count'] = 1
clean_df.sort_values('year', ascending = True)
```

```python
# now we can break it down to how many uniques were released each year

unique = clean_df.groupby(['year']).sum().reset_index()
unique

set_per_year = unique.groupby(['year']).sum().reset_index()[['year', 'count']]
set_per_year

for index, row in set_per_year.iterrows():
    print(row['year'], row['count'])
```

## MERGE DATA

#### join the two datasets to answer questions

```python
sets_themes = pd.merge(sets, themes, how = 'left', left_on = 'theme_id', right_on = 'id', suffixes = ('_sets', '_themes'))
sets_themes.drop(columns = 'img_url', inplace = True)
```

```python
sets_themes
```

## Answering questions

#### What is the total number of parts per theme

```python
parts_per_theme = sets_themes.groupby('name_themes').sum().reset_index()[['name_themes','name_sets','num_parts']]
parts_per_theme = parts_per_theme.sort_values('num_parts', ascending = False)
```

```python
parts_per_theme
```

#### what is the total number of parts per year

```python
parts_per_year = sets_themes.groupby('year').sum().reset_index()[['year','name_sets','num_parts']]
parts_per_year = parts_per_year.sort_values('num_parts', ascending = False)
```

```python
parts_per_year
```

#### how many sets were created in each century in the dataset

```python
years = sets_themes['year'].unique()
years.sort()
years
```

```python
sets_themes['century'] = np.where(sets_themes['year'] < 2001, '20th Century', '21st Century')
```

```python
# for this question, we are counting the total number of sets, and not the sum of parts
sets_per_century = sets_themes.groupby('century').count().reset_index()
```

```python
sets_per_century[['century', 'set_num']]
```

#### what percentage of sets of sets ever released in the 21st century were train themed?

```python
century21 = sets_per_century['set_num'][1]
century21
```

```python
trains = sets_themes[sets_themes['name_themes'].str.contains('Train')]
trains.shape[0]
```

```python
print((trains.shape[0]/century21)*100)
```
